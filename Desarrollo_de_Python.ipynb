# CARGA Y EXPLORACIÓN INICIAL DE DATOS

```python
print("="*80)
print("CARGANDO DATASET DE ENFERMEDADES CARDÍACAS")
print("="*80)

# URL del dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"

# Nombres de las columnas
columnas = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
            'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']

# Cargar datos
df_original = pd.read_csv(url, names=columnas, na_values='?')

print(f"\n✅ Datos cargados exitosamente")
print(f"   Dimensiones: {df_original.shape[0]} filas × {df_original.shape[1]} columnas")
print(f"   Columnas: {list(df_original.columns)}")
```

# EXPLORACION INICIAL

```python
print("="*80)
print("EXPLORACIÓN INICIAL DE DATOS")
print("="*80)

print("\n Primeras 10 filas del dataset:")
print("-"*80)
display(df_original.head(10))

print("\n Información general del dataset:")
print("-"*80)
df_original.info()

print("\n Estadísticas descriptivas:")
print("-"*80)
display(df_original.describe().round(2))

print("\n Análisis de valores nulos:")
print("-"*80)
nulos = df_original.isnull().sum()
print(nulos[nulos > 0])
print(f"\nTotal de valores nulos: {df_original.isnull().sum().sum()}")

print("\n Análisis de duplicados:")
print("-"*80)
duplicados = df_original.duplicated().sum()
print(f"Total de registros duplicados: {duplicados}")
```

# LIMPIEZA DE DATOS

```python
print("="*80)
print("LIMPIEZA Y PREPROCESAMIENTO DE DATOS")
print("="*80)

# Crear copia para no modificar el original
df_clean = df_original.copy()
filas_iniciales = len(df_clean)

print(f"\n Registros iniciales: {filas_iniciales}")
print("-"*80)

# 1. Eliminar valores nulos
nulos_antes = df_clean.isnull().sum().sum()
df_clean = df_clean.dropna()
nulos_eliminados = nulos_antes - df_clean.isnull().sum().sum()
print(f"\n Valores nulos eliminados: {nulos_eliminados}")
print(f"   Razón: Los valores nulos impiden análisis estadístico preciso")
print(f"   Registros después: {len(df_clean)}")

# 2. Eliminar duplicados
duplicados = df_clean.duplicated().sum()
df_clean = df_clean.drop_duplicates()
print(f"\n Registros duplicados eliminados: {duplicados}")
print(f"   Razón: Evitar conteo múltiple de pacientes")
print(f"   Registros después: {len(df_clean)}")

# 3. Validar rangos de edad (18-100 años)
edad_invalida = ((df_clean['age'] < 18) | (df_clean['age'] > 100)).sum()
df_clean = df_clean[(df_clean['age'] >= 18) & (df_clean['age'] <= 100)]
print(f"\n Registros con edad inválida eliminados: {edad_invalida}")
print(f"   Razón: Edades fuera del rango normal humano adulto (18-100)")
print(f"   Registros después: {len(df_clean)}")

# 4. Eliminar colesterol = 0 (biológicamente imposible)
colesterol_cero = (df_clean['chol'] == 0).sum()
df_clean = df_clean[df_clean['chol'] > 0]
print(f"\n Registros con colesterol = 0 eliminados: {colesterol_cero}")
print(f"   Razón: El colesterol en 0 es biológicamente imposible")
print(f"   Registros después: {len(df_clean)}")

# 5. Eliminar presión arterial = 0
presion_cero = (df_clean['trestbps'] == 0).sum()
df_clean = df_clean[df_clean['trestbps'] > 0]
print(f"\n Registros con presión arterial = 0 eliminados: {presion_cero}")
print(f"   Razón: Presión 0 indica error de medición")
print(f"   Registros después: {len(df_clean)}")

# Resumen final
print("\n" + "="*80)
print("RESUMEN DE LIMPIEZA")
print("="*80)
print(f"Registros iniciales: {filas_iniciales}")
print(f"Registros finales: {len(df_clean)}")
print(f"Registros eliminados: {filas_iniciales - len(df_clean)}")
print(f"Porcentaje conservado: {(len(df_clean)/filas_iniciales)*100:.2f}%")

# Mostrar dataset limpio
print("\n Dataset limpio (primeras 10 filas):")
display(df_clean.head(10))
```

# FILTRACION DE DATOS

```python
print("="*80)
print("FILTRACIÓN DE DATOS - SUBGRUPOS ESPECÍFICOS")
print("="*80)
print("\nEsta sección filtra el dataset en múltiples subgrupos para análisis detallado")
print("-"*80)

print("\n FILTRO 1: PACIENTES MAYORES DE 50 AÑOS")
print("-"*80)

mayores_50 = df_clean[df_clean['age'] > 50].copy()

print(f"Total de pacientes mayores de 50 años: {len(mayores_50)}")
print(f"Porcentaje del dataset: {(len(mayores_50)/len(df_clean))*100:.2f}%")
print(f"Edad promedio: {mayores_50['age'].mean():.1f} años")
print(f"Con enfermedad cardíaca: {mayores_50['target'].sum()}")
print(f"Tasa de enfermedad: {(mayores_50['target'].mean()*100):.2f}%")

print("\n Estadísticas del grupo:")
display(mayores_50[['age', 'chol', 'trestbps', 'target']].describe().round(2))

print("\n FILTRO 2: MUJERES CON HIPERTENSIÓN (Presión > 140 mmHg)")
print("-"*80)

mujeres_hipertension = df_clean[
    (df_clean['sex'] == 0) & (df_clean['trestbps'] > 140)
].copy()

print(f"Total de mujeres con hipertensión: {len(mujeres_hipertension)}")
print(f"Porcentaje del total de mujeres: {(len(mujeres_hipertension)/len(df_clean[df_clean['sex']==0]))*100:.2f}%")
print(f"Edad promedio: {mujeres_hipertension['age'].mean():.1f} años")
print(f"Presión promedio: {mujeres_hipertension['trestbps'].mean():.1f} mmHg")
print(f"Con enfermedad cardíaca: {mujeres_hipertension['target'].sum()}")
print(f"Tasa de enfermedad: {(mujeres_hipertension['target'].mean()*100):.2f}%")

print("\n📋 Primeras 5 filas del subgrupo:")
display(mujeres_hipertension[['age', 'sex', 'trestbps', 'chol', 'target']].head())

print("\n FILTRO 3: HOMBRES CON COLESTEROL ALTO (> 240 mg/dl)")
print("-"*80)

hombres_colesterol_alto = df_clean[
    (df_clean['sex'] == 1) & (df_clean['chol'] > 240)
].copy()

print(f"Total de hombres con colesterol alto: {len(hombres_colesterol_alto)}")
print(f"Porcentaje del total de hombres: {(len(hombres_colesterol_alto)/len(df_clean[df_clean['sex']==1]))*100:.2f}%")
print(f"Edad promedio: {hombres_colesterol_alto['age'].mean():.1f} años")
print(f"Colesterol promedio: {hombres_colesterol_alto['chol'].mean():.1f} mg/dl")
print(f"Con enfermedad cardíaca: {hombres_colesterol_alto['target'].sum()}")
print(f"Tasa de enfermedad: {(hombres_colesterol_alto['target'].mean()*100):.2f}%")

print("\n FILTRO 4: PACIENTES JÓVENES CON ENFERMEDAD CARDÍACA (< 45 años)")
print("-"*80)

jovenes_enfermos = df_clean[
    (df_clean['age'] < 45) & (df_clean['target'] == 1)
].copy()

print(f"Total de pacientes jóvenes con enfermedad: {len(jovenes_enfermos)}")
print(f"Edad promedio: {jovenes_enfermos['age'].mean():.1f} años" if len(jovenes_enfermos) > 0 else "No hay datos")
print(f"Distribución por sexo:")
if len(jovenes_enfermos) > 0:
    print(f"  - Mujeres: {len(jovenes_enfermos[jovenes_enfermos['sex']==0])}")
    print(f"  - Hombres: {len(jovenes_enfermos[jovenes_enfermos['sex']==1])}")
    display(jovenes_enfermos[['age', 'sex', 'chol', 'trestbps', 'cp']].head())

print("\n FILTRO 5: PACIENTES DE ALTO RIESGO (Múltiples factores)")
print("-"*80)
print("Criterios: Edad > 55, Colesterol > 240, Presión > 140")

alto_riesgo = df_clean[
    (df_clean['age'] > 55) &
    (df_clean['chol'] > 240) &
    (df_clean['trestbps'] > 140)
].copy()

print(f"\nTotal de pacientes de alto riesgo: {len(alto_riesgo)}")
print(f"Porcentaje del dataset: {(len(alto_riesgo)/len(df_clean))*100:.2f}%")
print(f"Con enfermedad cardíaca: {alto_riesgo['target'].sum()}")
print(f"Tasa de enfermedad: {(alto_riesgo['target'].mean()*100):.2f}%")

if len(alto_riesgo) > 0:
    print(f"\nCaracterísticas del grupo de alto riesgo:")
    print(f"  - Edad promedio: {alto_riesgo['age'].mean():.1f} años")
    print(f"  - Colesterol promedio: {alto_riesgo['chol'].mean():.1f} mg/dl")
    print(f"  - Presión promedio: {alto_riesgo['trestbps'].mean():.1f} mmHg")
    print(f"  - Distribución por sexo: {alto_riesgo['sex'].value_counts().to_dict()}")

print("\n FILTRO 6: PACIENTES CON FRECUENCIA CARDÍACA MÁXIMA BAJA (< 120 bpm)")
print("-"*80)

freq_baja = df_clean[df_clean['thalach'] < 120].copy()

print(f"Total de pacientes con frecuencia cardíaca baja: {len(freq_baja)}")
print(f"Porcentaje del dataset: {(len(freq_baja)/len(df_clean))*100:.2f}%")
print(f"Frecuencia promedio: {freq_baja['thalach'].mean():.1f} bpm")
print(f"Con enfermedad cardíaca: {freq_baja['target'].sum()}")
print(f"Tasa de enfermedad: {(freq_baja['target'].mean()*100):.2f}%")

print("\n FILTRO 7: ANÁLISIS POR TIPO DE DOLOR DE PECHO")
print("-"*80)

tipos_dolor = {
    0: 'Angina típica',
    1: 'Angina atípica',
    2: 'Dolor no anginoso',
    3: 'Asintomático'
}

print("Distribución por tipo de dolor de pecho:\n")
for tipo, nombre in tipos_dolor.items():
    grupo_dolor = df_clean[df_clean['cp'] == tipo]
    tasa_enfermedad = (grupo_dolor['target'].mean() * 100)
    print(f"  Tipo {tipo} ({nombre}):")
    print(f"    - Pacientes: {len(grupo_dolor)}")
    print(f"    - Con enfermedad: {grupo_dolor['target'].sum()}")
    print(f"    - Tasa: {tasa_enfermedad:.2f}%\n")

print("\n FILTRO 8: PACIENTES CON GLUCEMIA EN AYUNAS ELEVADA (fbs = 1)")
print("-"*80)

glucemia_elevada = df_clean[df_clean['fbs'] == 1].copy()

print(f"Total de pacientes con glucemia elevada: {len(glucemia_elevada)}")
print(f"Porcentaje del dataset: {(len(glucemia_elevada)/len(df_clean))*100:.2f}%")
print(f"Con enfermedad cardíaca: {glucemia_elevada['target'].sum()}")
print(f"Tasa de enfermedad: {(glucemia_elevada['target'].mean()*100):.2f}%")
print(f"Edad promedio: {glucemia_elevada['age'].mean():.1f} años")

print("\n" + "="*80)
print(" TABLA COMPARATIVA DE SUBGRUPOS FILTRADOS")
print("="*80)

resumen_filtros = pd.DataFrame({
    'Subgrupo': [
        'Mayores 50 años',
        'Mujeres hipertensas',
        'Hombres colesterol alto',
        'Jóvenes enfermos (<45)',
        'Alto riesgo (múltiple)',
        'Frecuencia cardíaca baja',
        'Glucemia elevada'
    ],
    'Total_Pacientes': [
        len(mayores_50),
        len(mujeres_hipertension),
        len(hombres_colesterol_alto),
        len(jovenes_enfermos),
        len(alto_riesgo),
        len(freq_baja),
        len(glucemia_elevada)
    ],
    'Con_Enfermedad': [
        mayores_50['target'].sum(),
        mujeres_hipertension['target'].sum(),
        hombres_colesterol_alto['target'].sum(),
        jovenes_enfermos['target'].sum(),
        alto_riesgo['target'].sum(),
        freq_baja['target'].sum(),
        glucemia_elevada['target'].sum()
    ],
    'Tasa_Enfermedad_%': [
        round(mayores_50['target'].mean() * 100, 2),
        round(mujeres_hipertension['target'].mean() * 100, 2),
        round(hombres_colesterol_alto['target'].mean() * 100, 2),
        round(jovenes_enfermos['target'].mean() * 100, 2) if len(jovenes_enfermos) > 0 else 0,
        round(alto_riesgo['target'].mean() * 100, 2) if len(alto_riesgo) > 0 else 0,
        round(freq_baja['target'].mean() * 100, 2),
        round(glucemia_elevada['target'].mean() * 100, 2)
    ],
    'Porcentaje_Dataset_%': [
        round((len(mayores_50)/len(df_clean)) * 100, 2),
        round((len(mujeres_hipertension)/len(df_clean)) * 100, 2),
        round((len(hombres_colesterol_alto)/len(df_clean)) * 100, 2),
        round((len(jovenes_enfermos)/len(df_clean)) * 100, 2),
        round((len(alto_riesgo)/len(df_clean)) * 100, 2),
        round((len(freq_baja)/len(df_clean)) * 100, 2),
        round((len(glucemia_elevada)/len(df_clean)) * 100, 2)
    ]
})

display(resumen_filtros)

print("\n GUARDANDO SUBGRUPOS FILTRADOS...")
print("-"*80)

# Guardar los subgrupos más importantes como archivos CSV
mayores_50.to_csv('subgrupo_mayores_50.csv', index=False)
mujeres_hipertension.to_csv('subgrupo_mujeres_hipertension.csv', index=False)
alto_riesgo.to_csv('subgrupo_alto_riesgo.csv', index=False)

print(" Subgrupos guardados exitosamente:")
print("   - subgrupo_mayores_50.csv")
print("   - subgrupo_mujeres_hipertension.csv")
print("   - subgrupo_alto_riesgo.csv")

print("\n" + "="*80)
print(" FILTRACIÓN DE DATOS COMPLETADA")
print("="*80)
```

# Creación de Grupos de Edad

```python
print("="*80)
print("CREACIÓN DE GRUPOS DE EDAD")
print("="*80)

# Crear grupos de edad
bins = [0, 40, 50, 60, 70, 100]
labels = ['<40', '40-50', '50-60', '60-70', '>70']
df_clean['grupo_edad'] = pd.cut(df_clean['age'], bins=bins, labels=labels)

print("\n Distribución por grupos de edad:")
print("-"*80)
distribucion_edad = df_clean['grupo_edad'].value_counts().sort_index()
print(distribucion_edad)

print("\n Porcentajes por grupo:")
print("-"*80)
porcentajes = (df_clean['grupo_edad'].value_counts(normalize=True) * 100).sort_index()
for grupo, porcentaje in porcentajes.items():
    print(f"   {grupo}: {porcentaje:.2f}%")

print("\n Grupos de edad creados exitosamente")
```

#  Filtración de Subgrupos

```python
print("="*80)
print("FILTRACIÓN DE SUBGRUPOS DE INTERÉS")
print("="*80)

# 1. Pacientes mayores de 50 años
mayores_50 = df_clean[df_clean['age'] > 50]
print(f"\n Pacientes mayores de 50 años: {len(mayores_50)}")
print(f"   Porcentaje del total: {(len(mayores_50)/len(df_clean))*100:.2f}%")
print(f"   Edad promedio: {mayores_50['age'].mean():.1f} años")

# 2. Mujeres (sex = 0)
mujeres = df_clean[df_clean['sex'] == 0]
print(f"\n Pacientes mujeres: {len(mujeres)}")
print(f"   Porcentaje del total: {(len(mujeres)/len(df_clean))*100:.2f}%")
print(f"   Con enfermedad cardíaca: {mujeres['target'].sum()}")
print(f"   Tasa de enfermedad: {(mujeres['target'].mean()*100):.2f}%")

# 3. Hombres (sex = 1)
hombres = df_clean[df_clean['sex'] == 1]
print(f"\n Pacientes hombres: {len(hombres)}")
print(f"   Porcentaje del total: {(len(hombres)/len(df_clean))*100:.2f}%")
print(f"   Con enfermedad cardíaca: {hombres['target'].sum()}")
print(f"   Tasa de enfermedad: {(hombres['target'].mean()*100):.2f}%")

# 4. Pacientes con colesterol alto (>240 mg/dl)
colesterol_alto = df_clean[df_clean['chol'] > 240]
print(f"\n Pacientes con colesterol alto (>240 mg/dl): {len(colesterol_alto)}")
print(f"   Porcentaje del total: {(len(colesterol_alto)/len(df_clean))*100:.2f}%")
print(f"   Colesterol promedio: {colesterol_alto['chol'].mean():.1f} mg/dl")

# 5. Pacientes hipertensos (presión >140 mmHg)
hipertensos = df_clean[df_clean['trestbps'] > 140]
print(f"\n Pacientes hipertensos (>140 mmHg): {len(hipertensos)}")
print(f"   Porcentaje del total: {(len(hipertensos)/len(df_clean))*100:.2f}%")
print(f"   Presión promedio: {hipertensos['trestbps'].mean():.1f} mmHg")

# Crear tabla comparativa
print("\n TABLA COMPARATIVA DE SUBGRUPOS:")
print("="*80)
comparativa = pd.DataFrame({
    'Subgrupo': ['Mayores 50', 'Mujeres', 'Hombres', 'Colesterol Alto', 'Hipertensos'],
    'N_Pacientes': [len(mayores_50), len(mujeres), len(hombres),
                    len(colesterol_alto), len(hipertensos)],
    'Porcentaje': [
        f"{(len(mayores_50)/len(df_clean))*100:.1f}%",
        f"{(len(mujeres)/len(df_clean))*100:.1f}%",
        f"{(len(hombres)/len(df_clean))*100:.1f}%",
        f"{(len(colesterol_alto)/len(df_clean))*100:.1f}%",
        f"{(len(hipertensos)/len(df_clean))*100:.1f}%"
    ]
})
display(comparativa)
```

#  Funciones Agregadas

```python
print("="*80)
print("ESTADÍSTICAS AGREGADAS POR VARIABLE")
print("="*80)

# Seleccionar columnas numéricas
columnas_numericas = df_clean.select_dtypes(include=[np.number]).columns.tolist()
columnas_numericas.remove('target')  # Remover target para análisis separado

# Calcular estadísticas agregadas
estadisticas = df_clean[columnas_numericas].agg([
    'count', 'mean', 'median', 'std', 'min', 'max'
]).T

print("\n Estadísticas descriptivas completas:")
print("-"*80)
display(estadisticas.round(2))

# Análisis por variable target
print("\n COMPARACIÓN: PACIENTES CON vs SIN ENFERMEDAD")
print("="*80)

for columna in ['age', 'chol', 'trestbps', 'thalach']:
    sin_enfermedad = df_clean[df_clean['target'] == 0][columna]
    con_enfermedad = df_clean[df_clean['target'] == 1][columna]

    print(f"\n{columna.upper()}:")
    print(f"  Sin enfermedad - Media: {sin_enfermedad.mean():.2f}, "
          f"Mediana: {sin_enfermedad.median():.2f}")
    print(f"  Con enfermedad - Media: {con_enfermedad.mean():.2f}, "
          f"Mediana: {con_enfermedad.median():.2f}")
    print(f"  Diferencia: {abs(con_enfermedad.mean() - sin_enfermedad.mean()):.2f}")
```

# Agrupaciones

```python
print("="*80)
print("AGRUPACIONES DE DATOS")
print("="*80)

# 1. Agrupación por grupo de edad
print("\n AGRUPACIÓN POR GRUPO DE EDAD:")
print("-"*80)
grupo_edad = df_clean.groupby('grupo_edad').agg({
    'age': 'count',
    'target': ['sum', 'mean']
}).round(3)
grupo_edad.columns = ['Total_Pacientes', 'Con_Enfermedad', 'Tasa_Enfermedad']
grupo_edad['Tasa_Porcentaje'] = (grupo_edad['Tasa_Enfermedad'] * 100).round(2)
display(grupo_edad)

# 2. Agrupación por sexo
print("\n AGRUPACIÓN POR SEXO:")
print("-"*80)
grupo_sexo = df_clean.groupby('sex').agg({
    'age': ['count', 'mean'],
    'target': ['sum', 'mean'],
    'chol': 'mean',
    'trestbps': 'mean'
}).round(2)
grupo_sexo.columns = ['Total', 'Edad_Promedio', 'Con_Enfermedad',
                      'Tasa_Enfermedad', 'Colesterol_Promedio', 'Presion_Promedio']
grupo_sexo.index = ['Mujer', 'Hombre']
display(grupo_sexo)

# 3. Agrupación por tipo de dolor de pecho
print("\n AGRUPACIÓN POR TIPO DE DOLOR DE PECHO (CP):")
print("-"*80)
grupo_dolor = df_clean.groupby('cp').agg({
    'age': 'count',
    'target': ['sum', 'mean']
}).round(3)
grupo_dolor.columns = ['Total_Pacientes', 'Con_Enfermedad', 'Tasa_Enfermedad']
display(grupo_dolor)

# 4. Agrupación combinada: Edad + Sexo
print("\n AGRUPACIÓN COMBINADA: GRUPO DE EDAD + SEXO:")
print("-"*80)
grupo_combinado = df_clean.groupby(['grupo_edad', 'sex']).agg({
    'age': 'count',
    'target': ['sum', 'mean']
}).round(3)
grupo_combinado.columns = ['Total', 'Con_Enfermedad', 'Tasa']
display(grupo_combinado)
```

# Tabla Consolidada

```python
print("="*80)
print("TABLA CONSOLIDADA FINAL")
print("="*80)

# Crear tabla consolidada
consolidado = df_clean.groupby(['grupo_edad', 'sex']).agg({
    'age': 'count',
    'target': ['sum', 'mean'],
    'chol': 'mean',
    'trestbps': 'mean',
    'thalach': 'mean'
}).round(2)

consolidado.columns = ['Total_Pacientes', 'Con_Enfermedad', 'Tasa_Enfermedad',
                       'Colesterol_Prom', 'Presion_Prom', 'Freq_Cardiaca_Prom']
consolidado = consolidado.reset_index()
consolidado['Sexo'] = consolidado['sex'].map({0: 'Mujer', 1: 'Hombre'})
consolidado = consolidado.drop('sex', axis=1)
consolidado['Tasa_Porcentaje'] = (consolidado['Tasa_Enfermedad'] * 100).round(1)

# Reordenar columnas
consolidado = consolidado[['grupo_edad', 'Sexo', 'Total_Pacientes',
                           'Con_Enfermedad', 'Tasa_Porcentaje',
                           'Colesterol_Prom', 'Presion_Prom', 'Freq_Cardiaca_Prom']]
consolidado.columns = ['Grupo_Edad', 'Sexo', 'Total', 'Con_Enfermedad',
                       'Tasa_%', 'Colesterol', 'Presión', 'Freq_Cardiaca']

print("\n TABLA CONSOLIDADA COMPLETA:")
print("="*80)
display(consolidado)

print("\n Guardando tabla consolidada...")
consolidado.to_csv('heart_disease_consolidado.csv', index=False)
print(" Tabla guardada como 'heart_disease_consolidado.csv'")
```
